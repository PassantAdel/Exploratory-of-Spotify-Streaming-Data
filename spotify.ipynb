### context :
##### This data is about spotify songs and best-knowing music streaming all over the world
* Artist: Name of artist.
* Sreams: Number of times the artist song have been streamed on spotify.
* Daily: The average number of  streams an artist receives in daily basis.
* As Lead: Number of streams to the artist when they are the lead or primary artist.
* As Solo: Number of streams generated by the artisit for their solo.
* As feature: number of streams the artist featured as a guest or collaborator on tracks by other artists.

## Libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("artists.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
cols_to_replace = ['Streams', 'As lead', 'Solo', 'As feature']
for col in cols_to_replace:
    df[col] = df[col].str.replace(',', '')
    df[col] = df[col].astype('float')
df.head()    
```

## Data visualization

## Heat Map

```python
plt.figure(figsize=(8,5))
heatmap = df.corr() 
sns.heatmap(data = heatmap)
plt.title('Correlation matrix')
plt.show()
```

```python
top10_streamed_artist = df.sort_values(by ='Streams', ascending=False).head(10)
top10_streamed_artist
```

```python
plt.figure(figsize=(8,5))
ax = sns.barplot(x ='Artist',y ='Streams',data =top10_streamed_artist,color='orange')
ax.tick_params(axis ='x' , rotation = 45) # adjust the tittle of each bar 
plt.ylabel('Total streams')
plt.title('Top 10 Most Sreamed Artists')
plt.show()
```

```python
top10_streamed_artists_daily = df.sort_values(by ='Daily', ascending= False).head(10)
plt.figure(figsize=(8,5))
ax = sns.barplot(x ='Artist',y ='Daily',data =top10_streamed_artists_daily,color='orange')
ax.tick_params(axis ='x' , rotation = 45) # adjust the tittle of each bar 
plt.ylabel('Daily Average Stream')
plt.title('Top 10 Most Sreamed Artists Daily')
plt.show()
```

```python
top10_solo_artists = df.sort_values(by ='Solo', ascending= False).head(10)
plt.figure(figsize=(8,5))
ax = sns.barplot(x ='Artist',y ='Solo',data =top10_solo_artists,color='orange')
ax.tick_params(axis ='x' , rotation = 45) # adjust the tittle of each bar 
plt.ylabel('Solo Streams')
plt.title('Top 10 Solo streamed Artists')
plt.show()
```
